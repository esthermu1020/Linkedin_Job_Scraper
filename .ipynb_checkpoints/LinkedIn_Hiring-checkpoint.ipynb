```python
class Google_Search:
    def __init__(self, api_key, cx):
        self.api_key = api_key
        self.cx = cx
        self.num_results = 10
        self.df = pd.DataFrame(columns=['title', 'link', 'description'])
        self.service = build("customsearch", "v1", developerKey=self.api_key)

    def search(self, query, must_words: List=[], num_results=10):
        self.num_results = num_results
        self.must_words = must_words
        result = self.service.cse().list(
            q=query,
            cx=self.cx,
            num=num_results
        ).execute()
        self.filter_useful_result(result)
        return result
    
    def filter_useful_result(self, result):
        for item in result["items"]:
            title = item['title']
            link = item['link']
            description = item['snippet'] if item['snippet'] is not None else ''
            for word in self.must_words:
                if word not in description:
                    continue
            self.store_result(title, link, description)
        return self.df


    def store_result(self, title, link, description):
        new_row = {
            'title': title,
            'link': link,
            'description': description
            }
        self.df = self.df.append(new_row, ignore_index=True)
        return self.df
```

```python
class LinkedIn_Person:
    # https://linkedin-api.readthedocs.io/en/latest/api.html
    def __init__(self, AK, SK):
        self.AK = AK
        self.SK = SK
        self.api = Linkedin(AK, SK)
        self.profile = None
        self.profile_nlp = None
        self.gcp = []
        self.aws = []
        self.azure = []
        self.notable_post = []
        self.API_KEY = None
        self.CX = None
        self.name = None

    def one_people_search(self, public_id):
        profile = self.api.get_profile(public_id)
        self.public_id = public_id
        self.profile = profile
        self.name = self.profile['firstName'] + ' ' + self.profile['lastName']
        self.profile_to_nlp()
        return profile
    
    def get_profile_posts(self, count = 100):
        posts = self.api.get_profile_posts(self.public_id, post_count = count)
        return posts

    def _mentions_annotate(self, stuff, name=''):
        profile_str = str(stuff).lower()
        aws_mentions = [f'[==={name}===]'+ sentence for sentence in profile_str.split('\n') if ' aws ' in sentence or ' amazon' in sentence] 
        gcp_mentions = [f'[==={name}===]'+ sentence  for sentence in profile_str.split('\n') if ' google ' in sentence or ' gcp ' in sentence]
        azure_mentions = [f'[==={name}===]'+ sentence  for sentence in profile_str.split('\n') if ' azure ' in sentence or ' microsoft ' in sentence\
                          or ' msft ' in sentence]
        return gcp_mentions, aws_mentions, azure_mentions 
    
    def google_search_annotate(self, API_KEY, CX, query=['QANTAS'], must_words=[]):
        self.API_KEY = API_KEY
        self.CX = CX
        GS = Google_Search(self.API_KEY, self.CX)
        for kw in ['aws', 'gcp', 'azure', 'amazon', 'google', 'microsoft', 'msft']:
            sub_query = query.copy()
            sub_query.append(self.name)
            sub_query.append(kw)
            result = GS.search(query, num_results=10, must_words = [self.name, kw] + must_words + query)
            for row in result:
                gcp, aws, azure = self._mentions_annotate(row['description'] + row['link'], name = 'google')
                self.gcp += gcp
                self.aws += aws
                self.azure += azure
        return f'There are {result.shape[0]} search result associated to {query}'
    
    def profile_mentions_annotate(self, public_id):
        profile = self.one_people_search(public_id)
        gcp, aws, azure = self._mentions_annotate(self.profile_nlp, name = 'profile')
        self.gcp += gcp
        self.aws += aws
        self.azure += azure
        return self.profile_nlp

    def post_mentions_annotate(self, count = 100):
        posts = self.get_profile_posts(count)
        for post in posts:
            if 'commentary' not in post.keys():
                continue
            gcp, aws, azure = self._mentions_annotate(post['commentary']['text']['text'], name = 'post')
            url = post['socialContent']['shareUrl']
            if gcp != []:
                if url not in self.notable_post:
                    # print('gcp', url)
                    self.notable_post.append(url)
                    gcp = [g + url for g in gcp]
            if aws!= []:
                if url not in self.notable_post:
                    # print('aws', url)
                    self.notable_post.append(url)
                    aws = [a + url for a in aws]
            if azure!= []:
                if url not in self.notable_post:
                    # print('azure', url)
                    self.notable_post.append(url)
                    azure = [a + url for a in azure]
            self.gcp += gcp
            self.aws += aws
            self.azure += azure
        return self.notable_post
    
    def profile_to_nlp(self):
        self.work_experience = ''
        if 'experience' in self.profile.keys():
            work_experience = \
                [experience['companyName'] 
                    + ' | start from ' 
                    + str(experience['timePeriod']['startDate']['year']) if 'timePeriod' in experience and 'startDate' in experience['timePeriod'] else 'NA'
                    +  " : " + experience['title']  
                    for experience in self.profile['experience']
                    ]
            self.work_experience = f'There are in total {len(work_experience)} work experiences in this profile' + \
                            '; '.join(work_experience)
        self.summary = ''
        if 'summary' in self.profile.keys():
            self.summary += self.profile['summary']
        self.headline = ''
        if 'headline' in self.profile.keys():
            self.summary += self.profile['headline']
        self.certification = ''
        if 'certification' in self.profile.keys():
            self.certification = '; '.join([c['name'] for c in self.profile['certifications']])
        self.skills = ''
        if 'skills' in self.profile.keys():
            self.skills = '; '.join([s['name'] for s in self.profile['skills']])
        self.profile_nlp = self.summary + self.work_experience + self.certification + self.skills
```

```python
class LinkedIn_Group:
    def __init__(self, AK, SK):
        self.AK = AK
        self.SK = SK
        self.df = pd.DataFrame(columns=['name', 'last_company', 'current_title', 
                                        'profile_description', 'notable_post', 'aws', 'gcp', 'azure'])

    def insert_row(self, profile, notable_post, person):
        last_company = person.profile['experience'][1]['companyName'] if len(person.profile['experience']) > 1 else 'NA'
        new_row = {
                'name': person.profile['firstName'] + ' ' + person.profile['lastName'],
                'last_company': last_company,
                'current_title': person.profile['experience'][0]['title'],
                'profile_description': profile,
                'notable_post': notable_post,
                'aws': person.aws,
                'gcp': person.gcp,
                'azure': person.azure
            }
        self.df = self.df.append(new_row, ignore_index=True)
        return self.df

    def group_search(self, public_ids, count = 100):
        for id in public_ids:
            person = LinkedIn_Person(self.AK, self.SK)
            profile = person.profile_mentions_annotate(id)
            notable_post = person.post_mentions_annotate(count)
            df = self.insert_row(profile, notable_post, person)
            display(df.tail(1))
        return df
```

```python
def get_company_urn_ids(company, keyword_title = '', limit=100, verbose = False):
    api = Linkedin(AK, SK)
    search_results = api.search_people(
        keyword_title = keyword_title,
        keyword_company=company,
        limit = limit
    )
    if verbose:
        print(search_results)
    urn_dic = {result['urn_id']: result['name'] for result in search_results}
    print(f'There are in total {len(urn_dic)} results retrived')
    return urn_dic
```
